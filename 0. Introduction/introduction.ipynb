```python
train_df['workclass'].value_counts()
```

```python
mode_category = train_df['workclass'].mode()[0]
train_df['workclass'].fillna(mode_category, inplace=True)
```

```python
train_df['occupation'].value_counts()
```

```python
# train_df['occupation'].fillna('Unknown', inplace=True)

mode_category = train_df['occupation'].mode()[0]
train_df['occupation'+'_imputed'] = np.where(train_df['occupation'].isnull(), 1, 0)
train_df['occupation'].fillna(mode_category, inplace=True)
```

```python
train_df['native-country'].value_counts()
```

```python
mode_category = train_df['native-country'].mode()[0]
# train_df['native-country'+'_imputed'] = np.where(train_df['native-country'].isnull(), 1, 0)
train_df['native-country'].fillna(mode_category, inplace=True)
```

```python
col_list = list(train_df.columns)
occi = col_list.index('occupation')
# occi = col_list.find('occupation')
occii = col_list.index('occupation_imputed')
# occii = col_list.find('occupation_imputed')

col_list.insert(occi+1, col_list.pop(occii))

train_df = train_df[col_list]
# train_df.reindex(columns=col_list)

train_df.head()
```
